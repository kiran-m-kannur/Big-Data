```python
import matplotlib.pyplot as plt
f1 = open('../Outputs/Logistic-Regression-Output','r')
f2 = open('../Outputs/SVM-Output','r')
lr=[]
svm=[]
for i in f1:
    lr.append(float(i))
for j in f2:
    svm.append(float(j))
x=range(1,11)

plt.title("Accuracy of Different Models")
plt.xlabel('Iteration')
plt.ylabel('Accuracy')

plt.xticks(x)
plt.plot(x, lr[:10], marker = 'o', c = 'g',label="Logistic Regression")
plt.plot(x, svm[:10], marker = 'o', c = 'r',label="Supprt Vector Machine")
plt.legend(loc="upper left")

plt.show()
```
